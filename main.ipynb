```python
import re
import jieba

def readText(fileName):
    with open(fileName,'r') as f:
        resultName = fileName.split('/')[-1]
        for i in f:
            i = i.strip() # 删除换行
            if '<p class="note">' in i:
                # 遇到文中的注释就跳过去
                pass
            else:
                # 删除所有的HTML标签及其内部的标记
                htmlText = re.findall(r"<.*?>",i,re.S)
                for s in htmlText:
                    i = i.replace(s,'')
                # 删除所有的注释序号
                note1 = re.findall(r"\[.*?\]",i,re.S)
                for s in note1:
                    i = i.replace(s,'')
                note2 = re.findall(r"〔.*?〕",i,re.S)
                for s in note2:
                    i = i.replace(s,'')
                note3 = re.findall(r"\(.*?\)",i,re.S)
                for s in note3:
                    i = i.replace(s,'')
                with open("purge/"+resultName,'a') as f:
                    f.write(i)
for i in range(1,120):
    i = str(i).zfill(3)
    readText(f'text/{i}.html')

```

```python
import jieba

def splitWrod(fileName):
    ls = []
    with open(fileName,'r') as f:
        for line in f:
            ls = jieba.lcut(line)
            ls += ls
        return ls
# def intergrateLs(ls):
#     intergrateResult = []

def counter(intergrateLs):
    d = {}
    for i in intergrateLs:
        d[i] = d.get(i,0) + 1
    ls = d.items()
    sortedLs = sorted(ls,key = lambda x:x[1],reverse = True)
    return sortedLs

def useUserWords():
    with open('userWords.txt','r') as f:
        for word in f:
            word = word.strip()
            jieba.add_word(word)

def main(f:int,l:int):
    useUserWords()
    intergrateLs = []
    for i in range(f,l):
        i = str(i).zfill(3)
        ls = splitWrod(f'purge/{i}.html')
        intergrateLs = intergrateLs + ls
    sortedLs = counter(intergrateLs)
    return sortedLs

# with open('result.txt')
if __name__ == '__main__':

    f = 39 # 开始回目
    l = 80 # 结束回目

    sortedLs = main(f,l)
    
    notPrint = ['，','。','：','、','“','“','‘','’','；','？','”','！']
    for i in range(len(sortedLs)):
        if sortedLs[i][0] in notPrint:
            pass
        else:
            # print(sortedLs[i])
            with open('中间40回统计结果.txt','a') as f:
                f.write(str(sortedLs[i]))
                f.write('\n')



```
